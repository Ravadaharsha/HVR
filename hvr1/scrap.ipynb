```python
import pandas as pd
D0=pd.read_csv('NIT_2020 1.csv')
D1=pd.read_csv('NIT_2021 1.csv')
D2=pd.read_csv('NIT_2022 1.csv')
D3=pd.read_csv('NIT_2023 1.csv')
D4=pd.read_csv('NIT_2024 2.csv')
```

```python
Data=pd.merge(D0,D1,how='inner',on=['Institute','Academic Program Name','Quota','Seat Type','Gender'])
Data=pd.merge(Data,D2,how='inner',on=['Institute','Academic Program Name','Quota','Seat Type','Gender'])
Data=pd.merge(Data,D3,how='inner',on=['Institute','Academic Program Name','Quota','Seat Type','Gender'])
Data=pd.merge(Data,D4,how='inner',on=['Institute','Academic Program Name','Quota','Seat Type','Gender'])
Data.to_csv('Final.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

def check(data):
    df = pd.Series(data)

    model = ARIMA(df, order=(1,1,0))  
    model_fit = model.fit()

    future_prediction = model_fit.forecast(steps=1)
    return future_prediction
```

```python
Data['Opening Rank(2020)']=Data['Opening Rank(2020)'].astype(int)
Data['Closing Rank(2020)']=Data['Closing Rank(2020)'].astype(int)
Data['Opening Rank(2021)']=Data['Opening Rank(2021)'].astype(int)
Data['Closing Rank(2021)']=Data['Closing Rank(2021)'].astype(int)
Data['Opening Rank(2022)']=Data['Opening Rank(2022)'].astype(int)
Data['Closing Rank(2022)']=Data['Closing Rank(2022)'].astype(int)
Data['Opening Rank(2023)']=Data['Opening Rank(2023)'].astype(int)
Data['Closing Rank(2023)']=Data['Closing Rank(2023)'].astype(int)
Data['Opening Rank(2024)']=Data['Opening Rank(2024)'].astype(int)
Data['Closing Rank(2024)']=Data['Closing Rank(2024)'].astype(int)
Data.dtypes
```

```python
def predict(x):
    op=[x['Opening Rank(2020)'],x['Opening Rank(2021)'],x['Opening Rank(2022)'],x['Opening Rank(2023)'],x['Opening Rank(2024)']]
    return int(check(op))
Data['opening Rank(2025)']=Data.apply(lambda x:predict(x),axis=1)
Data
```

```python
def predict(x):
    op=[x['Closing Rank(2020)'],x['Closing Rank(2021)'],x['Closing Rank(2022)'],x['Closing Rank(2023)'],x['Closing Rank(2024)']]
    return int(check(op))
Data['closing Rank(2025)']=Data.apply(lambda x:predict(x),axis=1)
Data
```
